```python
# fill null values based on dtype of the column
def fill_null(col):
    if col.dtype == 'object':
        return col.fillna('Unknown')
    elif col.dtype == 'int64':
        return col.fillna(0)
    elif col.dtype == 'float64':
        return col.fillna(0)

ipl.apply(fill_null)
```
